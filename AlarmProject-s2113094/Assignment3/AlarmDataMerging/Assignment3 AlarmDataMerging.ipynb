```python
ADL['ExternalID'] = ADL['ExternalID'].astype('int16').astype('str')
ADL_cl = ADL
ADL_cl
ADL_cl = ADL_cl.rename(columns={'AlarmServiceName':'ServiceName'})
#ADL_cl = ADL_cl[['ExternalID','ServiceName','Building','City','CareServiceType','Geog','HouseType','StartTime']]
ADL_cl['StartTime'] = pd.to_datetime(ADL_cl['StartTime'], infer_datetime_format=True)
```
