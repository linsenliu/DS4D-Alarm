```python
import pandas as pd
import numpy as np
```

```python
east3 = pd.read_csv('east3_sample_result.csv')
```

```python
east3
```

```python
east3_s = east3.groupby(['hour', 'ServiceName']).size().reset_index(name='ServiceCounts')
service = east3_s.pivot_table('ServiceCounts', ['hour'], 'ServiceName')
service.reset_index(drop=False, inplace=True)
service = service.reindex(['hour', 'General Alarm', 'Pull Cord Alarm', 'Wrist Alarm'], axis=1)
service = service.fillna(0)
service.to_csv('east3-service.csv',index=False,sep=',')
```

```python
east3_g = east3.groupby(['hour', 'Gender']).size().reset_index(name='GenderCounts')
gender = east3_g.pivot_table('GenderCounts', ['hour'], 'Gender')
gender.reset_index(drop=False, inplace=True)
gender = gender.reindex(['hour', 'F', 'M'], axis=1)
gender = gender.fillna(0)
gender.to_csv('east3-gender.csv',index=False,sep=',')
```

```python
east3_a = east3.groupby(['hour', 'AgeGroup']).size().reset_index(name='AgeCounts')
Age = east3_a.pivot_table('AgeCounts', ['hour'], 'AgeGroup')
Age.reset_index(drop=False, inplace=True)
Age = Age.reindex(['hour', '<60', '>60'], axis=1)
Age = Age.fillna(0)
Age.to_csv('east3-age.csv',index=False,sep=',')
```

```python
east3_d = east3.groupby(['hour', 'DisablingConditions']).size().reset_index(name='DisCounts')
Dis = east3_d.pivot_table('DisCounts', ['hour'], 'DisablingConditions')
Dis.reset_index(drop=False, inplace=True)
Dis = Dis.reindex(['hour', 'Physically Affected', 'Physically Not Affected'], axis=1)
Dis = Dis.fillna(0)
Dis.to_csv('east3-dis.csv',index=False,sep=',')
```
